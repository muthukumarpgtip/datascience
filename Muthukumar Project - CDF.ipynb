```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
fbdf = pd.read_csv("https://raw.githubusercontent.com/muthukumarpgtip/datascience/main/facebook_data.csv")
print("The dimension of the table is: ", fbdf.shape)
print(fbdf)
fbdf.head()
```

```python
#Summary of Data Set
print("Summary\n")
print('Rows     :',fbdf.shape[0])
print('Columns  :',fbdf.shape[1])
print('\nFeatures :\n     :',fbdf.columns.tolist())
print('\nMissing values    :',fbdf.isnull().values.sum())
print('\nUnique values :  \n',fbdf.nunique())
print('\nShape of Facebook Data Set: ',fbdf.shape)
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns
total = fbdf.isnull().sum().sort_values(ascending=False)
percent = (fbdf.isnull().sum()/fbdf.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
f, ax = plt.subplots(figsize=(15, 6))
plt.xticks(rotation='90')
sns.barplot(x=missing_data.index, y=missing_data['Percent'])
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
missing_data.head(9000)
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns
f,ax=plt.subplots(1,2,figsize=(18,8))
fbdf['gender'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Gender')
ax[0].set_ylabel('Count')
sns.countplot('gender',data=fbdf,ax=ax[1],order=fbdf['gender'].value_counts().index)
ax[1].set_title('Gender')
plt.show()
```
